```python
# I used these PyTorch models for CIFAR10 from this GitHub repository: https://github.com/huyvnphan/PyTorch_CIFAR10
# I made a few modifications so that these can work in .ipynb notebooks in Colab (see my repository)
# The state_dicts directory contains the pretrained weights for these models
# You could theoretically use any image classification model you like by modifying the code below and the Cifar10Model class

import torch
from robustbench.utils import load_model
import os

# print(os.getcwd())
# os.chdir("PyTorch_CIFAR10")
# print(os.getcwd())

# if os.path.isdir("state_dicts") is False:
#   print("need to rebuild state_dicts")
#   import data
#   import train
#   ! python train.py --download_weights 1

# from googlenet import googlenet
# from mobilenetv2 import mobilenet_v2
# from densenet import densenet121

device = 'cuda' if torch.cuda.is_available() else 'cpu'


class Cifar10Model:
    def __init__(self):
        # self.models = [googlenet, mobilenet_v2, densenet121]
        # self.model = self.models[idx](pretrained=True)
        # self.models = ["Gowal2021Improving_28_10_ddpm_100m",
                      #  "Gowal2020Uncovering_28_10_extra", "Standard"]
        # self.l_norm = ["Linf", "Linf", "Linf"]
        self.model = load_model(model_name='Standard', dataset='cifar10', threat_model='Linf')
        # self.model = load_model(model_name='Cui2023Decoupled_WRN-28-10', dataset='cifar10', threat_model='Linf')
        # self.model = load_model(model_name='Wong2020Fast', dataset='cifar10', threat_model='Linf')
        self.model.to(device)
        self.model.eval()

        self.max = 0
        self.counter = 0

    def enough(self):
        return True if self.counter >= self.max else False

    def set_maximum(self, max):
        self.max = max

    def predict(self, x):
        return self.model(x)

    def __call__(self, x):
        self.counter += 1
        return self.model(x)

    def zero_grad(self):
        self.model.zero_grad()
```
